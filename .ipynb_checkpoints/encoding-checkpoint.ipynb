# Encoding message in an image

````python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def modify_selected_channels(pixel_rgb, message_bits, bits_per_channel, channels_to_encode):
    expected_length = len(channels_to_encode) * bits_per_channel
    message_bits = message_bits.ljust(expected_length, '0')

    modified_pixel = list(pixel_rgb)
    bit_idx = 0

    for channel in channels_to_encode:
        bits = message_bits[bit_idx:bit_idx + bits_per_channel]
        bit_idx += bits_per_channel

        channel_idx = 'RGB'.index(channel)
        value = modified_pixel[channel_idx]

        # mask to clear the LSBs
        mask = 0xFF ^ ((1 << bits_per_channel) - 1)
        cleared = value & mask

        # Insert new bits
        modified_pixel[channel_idx] = cleared | int(bits, 2)

    return tuple(modified_pixel)


def encode_message_in_image(img_path, out_path, message, start_position=(0, 0), gap=0,
                            channels='RGB', num_bits=1, delimiter_start='#', delimiter_end='#',
                            horizontal=1):
    # checking channel input
    valid_channels = set('RGB')
    input_channels = set(channels.upper())
    if not input_channels.issubset(valid_channels):
        print("❌ Invalid channel input. Use only 'R', 'G', 'B'.")
        return

    # checking num_bits
    if not (1 <= num_bits <= 8):
        print("❌ Invalid num_bits. num_bits must be between 1 and 8.")
        return

    # checking horizontal input
    if horizontal not in [0, 1]:
        print("❌ Invalid horizontal value. It should be either 0 or 1.")
        return

    img = np.array(Image.open(img_path))
    rows, cols, _ = img.shape

    # checking start_position
    r_start, c_start = start_position
    if r_start >= rows or c_start >= cols:
        print(f"❌ Invalid start_position. The start position {start_position} exceeds image dimensions ({rows}, {cols}).")
        return

    channels = channels.upper()

    # Converting message to binary
    binary_message = ''.join(format(ord(c), '08b') for c in message)

    #converting delimiters to binary
    if delimiter_start:
        binary_start = ''.join(format(ord(c), '08b') for c in delimiter_start)
        binary_message = binary_start + binary_message

    if delimiter_end:
        binary_end = ''.join(format(ord(c), '08b') for c in delimiter_end)
        binary_message += binary_end

    # Calculate bits per pixel
    bits_per_pixel = len(channels) * num_bits

    # Calculate available pixels
    total_pixels = ((rows - r_start - 1) * cols + (cols - c_start)) if horizontal else ((cols - c_start - 1) * rows + (rows - r_start))
    available_pixels = total_pixels // (gap + 1) + (1 if total_pixels % (gap + 1) != 0 else 0)
    available_bits = available_pixels * bits_per_pixel
    max_characters = available_bits // 8

    # Checking if the message fits
    if len(binary_message) > available_bits:
        print(f"❌ Error: Message is too large to fit in the image. Max bits available: {available_bits}")
        print(f"   ➤ Message length (in bits): {len(binary_message)}")
        return

    # Pad binary message to match required length
    if len(binary_message) % bits_per_pixel != 0:
        binary_message = binary_message.ljust(
            ((len(binary_message) // bits_per_pixel) + 1) * bits_per_pixel,
            '0'
        )

    # FIXED Encoding process
    bit_pointer = 0

    if horizontal:
        # Horizontal traversal - FIXED logic
        for r in range(r_start, rows):
            # Start from c_start only for the first row, 0 for subsequent rows
            start_col = c_start if r == r_start else 0
            
            for c in range(start_col, cols, gap + 1):
                if bit_pointer >= len(binary_message):
                    break

                pixel = img[r, c]
                message_bits = binary_message[bit_pointer: bit_pointer + bits_per_pixel]
                modified_pixel = modify_selected_channels(
                    pixel, message_bits, num_bits, channels
                )

                img[r, c] = modified_pixel
                bit_pointer += bits_per_pixel

            if bit_pointer >= len(binary_message):
                break

    else:  # Vertical encoding - FIXED logic
        for c in range(c_start, cols):
            # Start from r_start only for the first column, 0 for subsequent columns
            start_row = r_start if c == c_start else 0
            
            for r in range(start_row, rows, gap + 1):
                if bit_pointer >= len(binary_message):
                    break

                pixel = img[r, c]
                message_bits = binary_message[bit_pointer: bit_pointer + bits_per_pixel]
                modified_pixel = modify_selected_channels(
                    pixel, message_bits, num_bits, channels
                )

                img[r, c] = modified_pixel
                bit_pointer += bits_per_pixel

            if bit_pointer >= len(binary_message):
                break

    # Save the modified image
    Image.fromarray(img).save(out_path)
    print(f"✅ Message successfully encoded and saved to '{out_path}'")


# Example usage - FIXED with proper LSB parameters
encode_message_in_image(
    img_path='images/man.jpg',               
    out_path='images/enc_man.png',           # Output as PNG
    message="hi guys iam from datascience"*100,  # Removed *100 for testing
    start_position=(0, 0),
    gap=1,                                   # Changed from 4 to 0 for sequential encoding
    channels='RGB',
    num_bits=8,                              # Changed from 8 to 1 for true LSB
    delimiter_start='#',
    delimiter_end='#',
    horizontal=1
)
````

````python
def extract_lsb_bits(pixel_rgb, bits_per_channel, channels_to_decode):
    """Extract LSB bits from pixel channels"""
    extracted_bits = ''
    
    for channel in channels_to_decode:
        channel_idx = 'RGB'.index(channel)
        value = pixel_rgb[channel_idx]
        
        # Extract the LSBs using mask
        mask = (1 << bits_per_channel) - 1
        bits = value & mask
        extracted_bits += format(bits, f'0{bits_per_channel}b')
    
    return extracted_bits


def decode_message_from_image(img_path, start_position=(0, 0), gap=0,
                             channels='RGB', num_bits=1, delimiter_start='#', delimiter_end='#',
                             horizontal=1, max_chars=1000):
    """Decode message from image using LSB technique - mirrors encoding logic exactly"""
    
    # Input validation (same as encoding)
    valid_channels = set('RGB')
    input_channels = set(channels.upper())
    if not input_channels.issubset(valid_channels):
        print("❌ Invalid channel input. Use only 'R', 'G', 'B'.")
        return ""

    if not (1 <= num_bits <= 8):
        print("❌ Invalid num_bits. num_bits must be between 1 and 8.")
        return ""

    if horizontal not in [0, 1]:
        print("❌ Invalid horizontal value. It should be either 0 or 1.")
        return ""
    
    img = np.array(Image.open(img_path))
    rows, cols, _ = img.shape
    channels = channels.upper()
    
    # Check start_position (same as encoding)
    r_start, c_start = start_position
    if r_start >= rows or c_start >= cols:
        print(f"❌ Invalid start_position. The start position {start_position} exceeds image dimensions ({rows}, {cols}).")
        return ""
    
    # Calculate bits per pixel (same as encoding)
    bits_per_pixel = len(channels) * num_bits
    
    # Extract binary message using SAME traversal logic as encoding
    extracted_bits = ''
    bit_pointer = 0
    
    if horizontal:
        # Horizontal traversal - EXACTLY same as encoding
        for r in range(r_start, rows):
            # Start from c_start only for the first row, 0 for subsequent rows
            start_col = c_start if r == r_start else 0
            
            for c in range(start_col, cols, gap + 1):
                if bit_pointer >= max_chars * 8 * len(channels):  # Safety limit
                    break
                    
                pixel = img[r, c]
                pixel_bits = extract_lsb_bits(pixel, num_bits, channels)
                extracted_bits += pixel_bits
                bit_pointer += bits_per_pixel
                
            if bit_pointer >= max_chars * 8 * len(channels):
                break
                
    else:  # Vertical traversal - EXACTLY same as encoding
        for c in range(c_start, cols):
            # Start from r_start only for the first column, 0 for subsequent columns
            start_row = r_start if c == c_start else 0
            
            for r in range(start_row, rows, gap + 1):
                if bit_pointer >= max_chars * 8 * len(channels):
                    break
                    
                pixel = img[r, c]
                pixel_bits = extract_lsb_bits(pixel, num_bits, channels)
                extracted_bits += pixel_bits
                bit_pointer += bits_per_pixel
                
            if bit_pointer >= max_chars * 8 * len(channels):
                break
    
    # Convert binary to text
    message = ''
    found_start = False
    
    # Process extracted bits in 8-bit chunks to form characters
    for i in range(0, len(extracted_bits) - 7, 8):
        byte = extracted_bits[i:i+8]
        if len(byte) == 8:
            try:
                char = chr(int(byte, 2))
                
                # Handle start delimiter
                if not found_start:
                    if delimiter_start and char == delimiter_start:
                        found_start = True
                        continue
                    elif not delimiter_start:
                        found_start = True
                
                # Add characters after start delimiter found
                if found_start:
                    if delimiter_end and char == delimiter_end:
                        break  # End delimiter found, stop
                    message += char
                        
            except ValueError:
                # Skip invalid characters
                continue
    
    return message


# Test both encoding and decoding with same parameters
print("Testing LSB Steganography - Encoding and Decoding")
print("=" * 60)

# Test message
test_message = "Hello World! This is a secccret message."
print(f"Original message: '{test_message}'")

# Encode the message
encode_message_in_image(
    img_path='images/man.jpg',
    out_path='images/test_encoded.png',
    message=test_message,
    start_position=(0, 0),
    gap=1,
    channels='RGB',
    num_bits=8,
    delimiter_start='#',
    delimiter_end='#',
    horizontal=1
)

# Decode the message using SAME parameters
decoded_message = decode_message_from_image(
    img_path='images/test_encoded.png',
    start_position=(0, 0),
    gap=1,
    channels='RGB',
    num_bits=8,
    delimiter_start='#',
    delimiter_end='#',
    horizontal=1,
    max_chars=100
)

print(f"Decoded message:  '{decoded_message}'")
print(f"Match: {decoded_message == test_message}")

if decoded_message == test_message:
    print("🎉 SUCCESS: Encoding and decoding work perfectly!")
else:
    print("❌ Issue detected - checking details...")
    print(f"Original length: {len(test_message)}")
    print(f"Decoded length: {len(decoded_message)}")
````

## Decoding Function

The `decode_message_from_image()` function extracts hidden messages from images using the same parameters as encoding:

### Key Features:
- **Same Parameters**: Uses identical parameters as encoding function
- **Mirror Logic**: Follows the exact same pixel traversal pattern as encoding
- **Input Validation**: Validates all parameters same as encoding
- **Flexible**: Supports all encoding options (horizontal/vertical, different channels, bit depths)

### Parameters:
- `img_path`: Path to the image containing hidden message
- `start_position`: Starting pixel position (row, col)
- `gap`: Gap between pixels (0 = sequential, 1 = every other pixel, etc.)
- `channels`: Which color channels to use ('RGB', 'R', 'G', 'B', 'RG', etc.)
- `num_bits`: Number of bits per channel (1-8, where 1 = LSB steganography)
- `delimiter_start` & `delimiter_end`: Message boundary markers
- `horizontal`: 1 for row-wise, 0 for column-wise traversal
- `max_chars`: Safety limit for extraction

### Usage:
```python
message = decode_message_from_image(
    img_path='encoded_image.png',
    start_position=(0, 0),
    gap=1,
    channels='RGB',
    num_bits=8,
    delimiter_start='#',
    delimiter_end='#',
    horizontal=1
)
```

**Important**: Use the exact same parameters that were used for encoding!
